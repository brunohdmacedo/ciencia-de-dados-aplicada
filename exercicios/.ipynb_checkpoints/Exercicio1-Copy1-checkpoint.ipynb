# Análise dos dados de Desmatamento da Amazônia Legal

A Amazônia Legal é uma área que engloba nove estados do Brasil pertencentes à bacia Amazônica e à área de ocorrência das vegetações amazônicas. O governo federal, reunindo regiões de idênticos problemas econômicos, políticos e sociais, com o intuito de melhor planejar o desenvolvimento social e econômico da região amazônica, instituiu o conceito de "Amazônia legal".

A atual área de abrangência da Amazônia Legal corresponde à totalidade dos estados do Acre, Amapá, Amazonas, Mato Grosso, Pará, Rondônia, Roraima e Tocantins e parte do estado do Maranhão (a oeste do meridiano de 44º de longitude oeste), perfazendo uma superfície de aproximadamente 5 217 423 quilômetros quadrados correspondente a cerca de 61% do território brasileiro. Sua população, entretanto, corresponde a 12,32% do total de habitantes do Brasil.

Nela, vivem em torno de 23 milhões de pessoas, segundo o censo 2010, distribuídas em 775 municípios, nos estados do Acre, Amapá, Amazonas, Mato Grosso, Pará, Rondônia, Roraima, Tocantins (98% da área do estado), Maranhão (79%) e Goiás (0,8%). Além de conter 20% do bioma cerrado, a região abriga todo o bioma Amazônia, o mais extenso dos biomas brasileiros, que corresponde a 1/3 das florestas tropicais úmidas do planeta, detém a mais elevada biodiversidade, o maior banco genético e 1/5 da disponibilidade mundial de água potável.[1]

Nos nove estados da Amazônia Legal, residem 55,9% da população indígena brasileira, ou seja, cerca de 250 mil pessoas, segundo o Sistema de Informação da Atenção à Saúde Indígena (SIASI) em abril de 2005 da Fundação Nacional de Saúde (FUNASA). A região abrange 24 dos 34 distritos sanitários especiais indígenas mantidos pela FUNASA e com uma grande diversidade étnica (cerca de 80 etnias).

```python
municipiosAM_df = municipios_df[municipios_df.codigo_uf ==13]
```

```python
municipiosAM_df.drop(['capital', 'codigo_uf'],axis=1, inplace=True)
```

Note que nos dados provenientes da tabela ***municipios.csv*** o nome dos municípios estão todos acentuados, no entanto na tabela amazonas os nomes estão sem acento. Logo é necessario remover os acentos dos nomes da tabela ***municipios.csv***.

```python
from unicodedata import normalize
# use esta funçao para remover acentos
def remove_acentos_python3(txt):
    return str(normalize('NFKD', txt).encode('ASCII', 'ignore'), 'utf-8')
```

```python
municipiosAM_df['nome'] = municipiosAM_df['nome'].apply(lambda nome: remove_acentos_python3(nome))
```

```python
municipiosAM_df.head()
```

Em seguida relacione as tabelas municipios e amazonas.

```python
merged_df = pd.merge(amazonas2014_df, municipiosAM_df, how='left', left_on=['Municípios'], right_on=['nome'])
```

```python
merged_df.head()
```

### Questão


Plote no mapa do Brasil indicadores  de desflorestamento para todos os municipios do amazonas.
Os indicadores devem ser relativos ao tamanho da floresta de cada município dando destaque aos municipios que com desflorestamento acima da média.

#### Dica

O desflorestamento relativo é dado pelas colunas ***Desflorestamento 2014*** / ***Floresta*** da tabela ***amazonas.csv***.
Note que a coluna ***Floresta*** não é numérica e contém ',' separando os milhares, sendo necessário remover a ',' de todas as entradas e logo após converter os valores para float.

Podem ser usadas todas as dicas relevantes sobre apresentação de figuras  nos textos *Making Data Meaningful* e *Ten Simple Rules for Better Figures*.

Para plotar sobre o mapa, pode-se usar a funções normais do matplotlib e ao fim chamar a função **mplleaflet.display()**.

Verificando tipo das colunas:

```python
merged_df.dtypes
```

```python
merged_df['Floresta'] = merged_df['Floresta'].apply(lambda x: float(str(x).replace(",","")))
merged_df['Hidrografia'] = merged_df['Hidrografia'].apply(lambda x: float(str(x).replace(",","")))
merged_df['Não Floresta'] = merged_df['Não Floresta'].apply(lambda x: float(str(x).replace(",","")))
merged_df['Pasto Limpo'] = merged_df['Pasto Limpo'].apply(lambda x: float(str(x).replace(",","")))
merged_df['Total Geral'] = merged_df['Total Geral'].apply(lambda x: float(str(x).replace(",","")))
```

```python
merged_df.dtypes
```

## Visualizando Indicadores de Desflorestamento

```python
merged_df.boxplot(column=['Floresta', 'Não Floresta','Total Geral'], figsize=(15,7))
```

```python
#calcula as informações do desflorestamento
merged_df['Desflorestamento Relativo'] = merged_df['Desflorestamento 2014']/merged_df['Floresta']

desfl_mean = merged_df['Desflorestamento Relativo'].mean()
desfl_min = merged_df['Desflorestamento Relativo'].min()
desfl_max = merged_df['Desflorestamento Relativo'].max()

merged_df['Desflorestamento Relativo Acima da Média'] = merged_df['Desflorestamento Relativo'].apply(lambda x: 1 if x > desfl_mean else 0)
```

```python
merged_df.head()
```

```python
merged_df.boxplot(column=['Desflorestamento Relativo'], figsize=(15,7))
```

```python
#codigo de plot
merged_df.plot.scatter(x='longitude',y='latitude', 
                       s=merged_df['Floresta']/(merged_df['Floresta'].min()),  
                       c='Desflorestamento Relativo Acima da Média', cmap='coolwarm', alpha=0.8,
                       figsize=(8,8))
# for lat,long in zip(merged_df.latitude, merged_df.longitude):
#     plt.scatter([long], [lat], marker='o', c='purple')

#transforma o plot em localizações no mapa
mplleaflet.display()
```

## Relação entre Mineração e Desmatamento (dados de 2014)

### Questão


Segundo os dados, a mineração nos municípios tem alguma relação com o desflorestamento acima da média?
Mostre suas conclusões por tabelas e plots no mapa.

### Calculando relaçoes entre desflorestamento e mineração

```python
merged_df[['Mineração', 'Desflorestamento Relativo']].corr()
```

```python
merged_df[['Mineração', 'Desflorestamento Relativo Acima da Média' ]].corr()
```

```python
merged_df[['Mineração', 'Desflorestamento Relativo Acima da Média' ]]\
        .sort_values(by='Mineração', ascending=False)[:15]\
        .groupby('Desflorestamento Relativo Acima da Média').describe()
```

### Visualizando Desflorestamento e Mineração

#### 1. Distribuição dos valores da coluna Mineração

```python
merged_df.boxplot(column=['Mineração'], figsize=(12,5))
```

#### 2. Distribuição dos valores da coluna Mineração vs. Desflorestamento Relativo

```python
fig, ax = plt.subplots(1,2, figsize=(16,5), sharey='all')
plt.subplots_adjust(wspace=0,hspace=0)

merged_df.sort_values(by='Desflorestamento Relativo', ascending=False)[16:]\
.plot.scatter('Desflorestamento Relativo', 'Mineração',  
                       s=60, alpha=0.6,ax=ax[0],
                     xlim=(merged_df['Desflorestamento Relativo'].min(), 
                           merged_df['Desflorestamento Relativo'].mean()-merged_df['Desflorestamento Relativo'].mean()/10),
             title='Desflorestamento Relativo Abaixo da Média')

merged_df.sort_values(by='Desflorestamento Relativo', ascending=False)[:16]\
.plot.scatter('Desflorestamento Relativo', 'Mineração', 
                       s=60, alpha=0.6, ax=ax[1],
              xlim=(merged_df['Desflorestamento Relativo'].mean()-merged_df['Desflorestamento Relativo'].mean()/10, 
                    merged_df['Desflorestamento Relativo'].max()+merged_df['Desflorestamento Relativo'].max()/10),
             title='Desflorestamento Relativo Acima da Média')
```

```python
fig, ax = plt.subplots(figsize=(16,5))
ax.set_xlabel('Desflorestamento Relativo')
merged_df.sort_values(by='Desflorestamento Relativo', ascending=False)\
.plot.scatter('Desflorestamento Relativo', 'Mineração',  s=40, alpha=0.7,
              xlim=(merged_df['Desflorestamento Relativo'].min()-0.0001,
                   merged_df['Desflorestamento Relativo'].max()+0.0001),
             c='Mineração', cmap='coolwarm', ax=ax)
```

#### 3. Distribuição dos valores da coluna Mineração vs. Desflorestamento Relativo Acima da Média

```python
merged_df[['Municípios', 'Mineração', 'Desflorestamento Relativo', 
           'Desflorestamento Relativo Acima da Média', ]]\
        .sort_values(by='Mineração', ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(15,4))
ax.set_xlabel('Mineração')
merged_df.plot.scatter('Mineração', 'Desflorestamento Relativo Acima da Média', 
                       s=50, alpha=0.7, ax=ax,
                       c='Mineração', cmap='coolwarm')
```

```python
merged_df[['Municípios', 'Mineração', 'Desflorestamento Relativo', 
           'Desflorestamento Relativo Acima da Média', ]]\
.sort_values(by='Desflorestamento Relativo', ascending=False)[:20]
```
