Carregamento e pré-processamento dos dados a serem analisados:

```python
#importação de pacotes
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#pacote para plotar no mapa
import mplleaflet
```

```python
#Importando dados do ibge de estados e municípios
amazonas2014_df = pd.read_csv("amazonas.csv")
estados_df = pd.read_csv("estados.csv")
municipios_df = pd.read_csv("municipios.csv")
```

Começaremos olhando a forma dos DataFrames criados:

```python
amazonas2014_df
```

```python
estados_df.head()
```

```python
municipios_df.head()
```

Após carregar os dados devemos associar as tabelas de estados e municípios e selecionar apenas os municípios do Amazonas:

```python
estados_df[estados_df.uf=='AM']
```

Criando um novo objeto DataFrame para armazenar os dados acerca dos municípios do Amazonas:

```python
municipiosAM_df = municipios_df[municipios_df.codigo_uf ==13]
```

```python
municipiosAM_df.head()
```

Note que nos dados provenientes da tabela ***municipios.csv*** o nome dos municípios estão todos acentuados, no entanto na tabela amazonas os nomes estão sem acento. Logo é necessario remover os acentos dos nomes da tabela ***municipios.csv***.

```python
from unicodedata import normalize
# use esta funçao para remover acentos
def remove_acentos_python3(txt):
    return str(normalize('NFKD', txt).encode('ASCII', 'ignore'), 'utf-8')
```

```python
municipiosAM_df['nome'] = municipiosAM_df['nome'].apply(lambda nome: remove_acentos_python3(nome))
```

```python
municipiosAM_df.head()
```

Em seguida relacionamos as tabelas municipios e amazonas em um novo DataFrame:

```python
merged_df = pd.merge(amazonas2014_df, municipiosAM_df, how='left', left_on=['Municípios'], right_on=['nome'])
```

```python
merged_df.head()
```

Visualizando Dataframe no mapa:

```python
#codigo de plot
merged_df.plot.scatter(x='longitude',y='latitude',  
                       c='yellow',
                       figsize=(8,8))
# for lat,long in zip(merged_df.latitude, merged_df.longitude):
#     plt.scatter([long], [lat], marker='o', c='purple')

#transforma o plot em localizações no mapa
mplleaflet.display(tiles='esri_aerial')
```

Verificando o tipo das colunas do DataFrame:

```python
merged_df.dtypes
```

Pode-se observar que existem colunas com dados numéricos que estão indicadas com o tipo *object*. Isso ocorre pois estas colunas estão com sua formatação no padrão de dados americandos (D,DDD.DD) para modificar isso basta eliminarmos a vírgula que divide os milhares dos números.

```python
merged_df['Floresta'] = merged_df['Floresta'].apply(lambda x: float(str(x).replace(",","")))
merged_df['Hidrografia'] = merged_df['Hidrografia'].apply(lambda x: float(str(x).replace(",","")))
merged_df['Não Floresta'] = merged_df['Não Floresta'].apply(lambda x: float(str(x).replace(",","")))
merged_df['Pasto Limpo'] = merged_df['Pasto Limpo'].apply(lambda x: float(str(x).replace(",","")))
merged_df['Total Geral'] = merged_df['Total Geral'].apply(lambda x: float(str(x).replace(",","")))
```

Verificando a alteração podemos observar agora que apenas as colunas que possuem strings permanecerão com o tipo *object*:

```python
merged_df.dtypes
```

## 1. Indicadores de Desflorestamento

#### 1.1 Desflorestamento Absoluto

Olhando para e extensão do desflorestamento em valores absolutos:

```python
merged_df[['Municípios', 'Desflorestamento 2014']]\
        .sort_values(by='Desflorestamento 2014', ascending=False)
```

Podemos utilizar um boxplot para comparar a distribuição do Desflorestamento 2014

```python
merged_df.boxplot(column=['Desflorestamento 2014'], figsize=(15,7))
```

#### 1.2 Desflorestamento Relativo

Com intuito de comparar taxas de desflorestamento nos diferentes municípios criaremos a variável *Desflorestamento Relativo*. No escopo dessa análise o desflorestamento relativo é dado pelas colunas ***Desflorestamento 2014*** / ***Floresta***, onde:

***Desflorestamento 2014*** - percentual de floresta desflorestada

***Floresta*** - área da floresta

```python
#calcula as informações do desflorestamento
merged_df['Desflorestamento Relativo'] = merged_df['Desflorestamento 2014']/merged_df['Floresta']

desfl_mean = merged_df['Desflorestamento Relativo'].mean()
desfl_min = merged_df['Desflorestamento Relativo'].min()
desfl_max = merged_df['Desflorestamento Relativo'].max()
```

Além disso, criaremos uma coluna indicará se um município possui *Desflorestamento Relativo Acima da Média*:

```python
merged_df['Desflorestamento Relativo Acima da Média'] = merged_df['Desflorestamento Relativo'].apply(lambda x: 1 if x > desfl_mean else 0)
```

```python
merged_df[['Municípios','Desflorestamento Relativo Acima da Média']].groupby('Desflorestamento Relativo Acima da Média').describe()
```

Visualizando o DataFrame com suas novas colunas:

```python
merged_df.head()
```

Observando a distribuição dos valores do Desflorestamento Relativo

```python
merged_df.boxplot(column=['Desflorestamento Relativo'], figsize=(15,7))
```

Como é de se esperar o Desflorestamento Relativo possui menor variancia em relação ao Desflorestamento Absoluto.

#### 1.3 Desflorestamento Relativo por Municípios

```python
#codigo de plot
merged_df.plot.scatter(x='longitude',y='latitude', 
                       s=merged_df['Floresta']/(merged_df['Floresta'].min()),  
                       c='Desflorestamento Relativo', cmap='coolwarm', alpha=0.9,
                       figsize=(8,8))
# for lat,long in zip(merged_df.latitude, merged_df.longitude):
#     plt.scatter([long], [lat], marker='o', c='purple')

#transforma o plot em localizações no mapa
mplleaflet.display()
```

No gráfico acima o tamanho dos marcadores representa o tamanho da *Floresta* ao passo que as cores indicam a instensidade do *Desflorestamento Relativo*.

#### 1.4 Desflorestamento Relativo Acima da Média por Municípios

```python
#codigo de plot
merged_df.plot.scatter(x='longitude',y='latitude', 
                       s=merged_df['Floresta']/(merged_df['Floresta'].min()),  
                       c='Desflorestamento Relativo Acima da Média', cmap='coolwarm', alpha=0.7,
                       figsize=(8,8))
# for lat,long in zip(merged_df.latitude, merged_df.longitude):
#     plt.scatter([long], [lat], marker='o', c='purple')

#transforma o plot em localizações no mapa
mplleaflet.display()
```

No gráfico acima o tamanho dos marcadores representa o tamanho da *Floresta* ao passo que as cores indicam a instensidade do *Desflorestamento Relativo Acima da Média*. Nesse contexto, a cor vermelha indica que a Floresta possui desflorestamento acima da média. 

## 2. Relação entre Mineração e Desmatamento (dados de 2014)

#### 2.1 Calculando relações entre desflorestamento e mineração

```python
merged_df[['Mineração', 'Desflorestamento Relativo']].corr()
```

```python
merged_df[['Mineração', 'Desflorestamento Relativo Acima da Média' ]].corr()
```

#### Observação 1: Correlação Negativa
A correlação negativa se dá quando $r < 0$, e nesse caso à medida que $x$ cresce, $y$ decresce (em média).

Em ambas tabelas acima que relacionam mineração e desflorestamento encontrou-se correlações negativas entre as colunas observadas.

```python
merged_df[['Mineração', 'Desflorestamento Relativo Acima da Média' ]]\
        .sort_values(by='Mineração', ascending=False)[:15]\
        .groupby('Desflorestamento Relativo Acima da Média').describe()
```

#### Observação 2: Distribuição Mineração vs Desflorestamento Relativo Acima da Média	
A distribuição é bastante parelha (8x7) e não permite inferir que a mineração provoque desflorestamento acima da média.

#### 2.2 Distribuição dos valores da coluna Mineração

```python
merged_df['Mineração'].count()
```

```python
merged_df.boxplot(column=['Mineração'], figsize=(12,5))
```

#### Observação 3: Poucos valores de mineração
Podemos observar que apenas quinze(de sessenta e dois) valores não são **NaN** e desses quinze, três destoam da média. A ausência de informação alcança 75,8% do conjunto de dados.

#### 2.3 Distribuição dos valores da coluna Mineração vs. Desflorestamento Relativo

Agrupando os dados por *Mineração*:

```python
merged_df[['Municípios', 'Mineração', 'Desflorestamento Relativo',
           'Desflorestamento Relativo Acima da Média' ]]\
        .sort_values(by='Mineração', ascending=False)
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,5), sharey='all')
plt.subplots_adjust(wspace=0,hspace=0)

merged_df.sort_values(by='Desflorestamento Relativo', ascending=False)[16:]\
.plot.scatter('Desflorestamento Relativo', 'Mineração',  
                       s=60, alpha=0.6,ax=ax[0],
                     xlim=(merged_df['Desflorestamento Relativo'].min(), 
                           merged_df['Desflorestamento Relativo'].mean()-merged_df['Desflorestamento Relativo'].mean()/10),
             title='Desflorestamento Relativo Abaixo da Média')

merged_df.sort_values(by='Desflorestamento Relativo', ascending=False)[:16]\
.plot.scatter('Desflorestamento Relativo', 'Mineração', 
                       s=60, alpha=0.6, ax=ax[1],
              xlim=(merged_df['Desflorestamento Relativo'].mean()-merged_df['Desflorestamento Relativo'].mean()/10, 
                    merged_df['Desflorestamento Relativo'].max()+merged_df['Desflorestamento Relativo'].max()/10),
             title='Desflorestamento Relativo Acima da Média')
```

Podemos utilizar um mapa de cores para auxiliar na distinção dos valores de *Mineração*:

```python
fig, ax = plt.subplots(figsize=(16,5))
ax.set_xlabel('Desflorestamento Relativo')
merged_df.sort_values(by='Desflorestamento Relativo', ascending=False)\
.plot.scatter('Desflorestamento Relativo', 'Mineração',  s=40, alpha=0.7,
              xlim=(merged_df['Desflorestamento Relativo'].min()-0.0001,
                   merged_df['Desflorestamento Relativo'].max()+0.0001),
             c='Mineração', cmap='coolwarm', ax=ax)
```

#### Observação 4: Nenhuma tendência relevante observada
Os gráficos de MineraçãoxDesflorestamento Relativo não indicam tendência de crescimento do desflorestamento com o crescimento da mineração.

E se olharmos apenas para os municípios que possuem *Desflorestamento Relativo Acima da Média*?
